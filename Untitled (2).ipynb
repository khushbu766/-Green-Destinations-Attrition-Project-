```python
# project green_destinations?
# Note :- 
#ðŸ“˜ Project Objective 
# The HR Director wants to understand:
# What percentage of employees have left the company
# ðŸ‘‰ This means calculating the Attrition Rate.
# (Attrition refers to the number of employees who have left the company.)
# She also wants to find out which factors might influence an employeeâ€™s decision to leave the company.
# For example:
# Age â€“ Does age affect whether employees stay or leave?
# Years at Company â€“ Do employees who have been with the company longer tend to stay?
# Income â€“ Are employees with lower income more likely to leave?


import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv("C:/Users/Khushboo/Downloads/greendestination.csv")
print(df)
```

```python
df.head()
```

```python
# Missing Values
df.isnull().sum()
```

```python
# Duplicate Rows
df = df.dropna()  # à¤¯à¤¾ df.fillna(method='ffill')
```

```python
df.duplicated().sum()
df = df.drop_duplicates()
```

```python
# Data Type Check
df.info()
```

```python
df['MonthlyIncome'] = df['MonthlyIncome'].astype(float)
```

```python
# Basic Info
df.describe()
```

```python
# Attrition Rate
attrition_rate = df['Attrition'].value_counts(normalize=True)['Yes'] * 100
print(f"Attrition Rate: {attrition_rate:.2f}%")
```

```python
# Relationship Check
import seaborn as sns
import matplotlib.pyplot as plt

sns.boxplot(x='Attrition', y='MonthlyIncome', data=df)
plt.title("Income vs Attrition")
plt.show()

sns.boxplot(x='Attrition', y='YearsAtCompany', data=df)
plt.title("Years at Company vs Attrition")
plt.show()
```
